```python
#  원리 

import random

print(random.random()) # 0~1 사이의 랜덤한 실수 출력
print(random.randint(1, 10)) # 1~10 사이의 랜덤한 정수 출력
print(random.choice([1, 2, 3, 4, 5])) # 리스트 안에 있는 값 중 랜덤하게 하나 출력

# balls -> shuffle

balls = list(range(1, 46)) # 1~45까지의 숫자를 리스트로 생성

for i in range(0,1000000): # 섞는 횟수를 지정할 수 있다는 장점이 있음
    temp = balls[0] # 0번째 값을 임시로 저장
    r_num = int(random.random() * len(balls))
    # 0 ~ 44 인덱스에 balls의 개수를 곱해서 0~44 사이의 랜덤한 정수 생성
    balls[0] = balls[r_num] # 0번째 값에 랜덤한 인덱스의 값을 대입
    balls[r_num] = temp # 최종 대입하여 자리 바꾸기(섞기)

print(balls) # 섞인 리스트 출력

print(balls[:6]) # 섞인 리스트에서 0~5번째 인덱스까지 출력 (6개)
```

```python
# 간단히 함수를 이용한 방법

# random 모듈 안 sample() 함수 사용
lotto = random.sample(range(1, 46), 6) # 1~45까지의 숫자 중 랜덤하게 6개를 뽑아 리스트로 반환
lotto.sort() # 원본을 정렬
print(lotto) # 정렬된 리스트 출력
```

```python
for i in range(10):
    lotto = random.sample(range(1, 46), 6)
    lotto.sort()
    print(f"{i}번째 로또 번호: {lotto}")
```

```python
def my_lotto_create(a):
    result = []
    for i in range(a):
        lotto = random.sample(range(1, 46), 6)
        lotto.sort()
        # print(f"{i+1}번째 로또 번호: {lotto}")
        result.append(tuple(lotto))
    return tuple(result)


result = my_lotto_create(3)
print(result)
```

```python
def dice_cast():
    d1 = random.randint(1,6)
    d2 = random.randint(1,6)
    return d1 + d2

# 딕셔너리로 각각 이름과 결과값을 저장 가능
person1_dice = {"name": "미치에다슌스케", "dice_value": 0}
person1_dice["dice_value"] = dice_cast()
person2_dice = {"name": "카미시라이시모네", "dice_value": 0}
person2_dice["dice_value"] = dice_cast()

# person1_dice = dice_cast();
# person2_dice = dice_cast();

if person1_dice["dice_value"] == person2_dice["dice_value"]:
    # print(f"비겼다. person1: {person1_dice}, person2: {person2_dice}")
    print(f"비겼다. 밋치: {person1_dice["dice_value"]}, 모네: {person2_dice["dice_value"]}")
elif person1_dice["dice_value"] > person2_dice["dice_value"]:
    # print(f"{person1_dice['name']} win. person1: {person1_dice}, person2: {person2_dice}")
    # print(f"{person1_dice['name']} win. person1: {person1_dice["dice_value"]}, person2: {person2_dice["dice_value"]}")
    print(f"{person1_dice['name']} win. 밋치: {person1_dice["dice_value"]}, 모네: {person2_dice["dice_value"]}")
else:
    print(f"{person2_dice['name']} win. 밋치: {person1_dice["dice_value"]}, 모네: {person2_dice["dice_value"]}")
```
