```python
def add(a, b):
    return a + b

# def - 함수를 정의하는 키워드
# add - 함수 이름
# (a, b) - 매개변수(Parameter)
# : - 함수 코드블록의 시작
# return - 함수종료와 return뒤에 값(결과 객체)를 호출한 곳(대입문이 위치)으로 반환
```

```python
def hello():
    """
    '안녕'을 출력하고 None 값을 반환하는 함수 
    """
    print("안녕")

result = hello()
# 함수호출은 표현식이다. 표현식은 값을 남긴다. 반환값이 없는 함수를 호출하면 None값을 남긴다.
print(result) # result에는 None값이 할당되어 있다.
```

```python
def calc(a, b):
    x = a + b
    y = a - b
    z = a * b
    return x , y, z # 3개의 값을 반환하는 것이 아니라 tuple로 묶어서 반환

r1 = calc(1,2)
print(r1)

r1, r2, r3 = calc(10,2) # 튜플 언패킹
print(r1)
print(r2)
print(r3)
```

```python
def fun1(a, b):
    print("매개변수 a의 인수: ", a)
    print("매개변수 b의 인수: ", b)

# 함수 호출시 인자값(인수)의 위치로 매개변수 값 결정
fun1(1, 2)
fun1(10, 20)
# fun1(100, 200, 300) # 매개변수보다 인자값이 적거나 많으면 오류 발생


# 함수 호출시 매개변수 이름을 지정하여 인자값(인수) 입력
fun1(a=100, b=200)
fun1(b=777, a=999) # 매개변수 이름을 지정하여 인수 입력시(키워드 인수) 순서는 상관없음.
fun1(b=777, 999) # 매개변수 이름을 지정하지 않은 인수는 위치로 결정되므로 오류 발생
```

```python
def func(x): # 불변데이터: 숫자, 문자, 튜플 
    x += 10
    print("함수 안:", x, id(x))

a = 10

func(a)

print("함수 밖:", a, id(a)) 
# 불변데이터니까 10이 아닌 20이라는 새로운 객체가 만들어지고, a는 여전히 10을 가리킨다.

# 함수 안: 20 140714686129672
# 함수 밖: 10 140714686129352
# 함수 안의 x만 변경, 함수 밖의 a는 그대로
# id()값이 다르다
```

```python
def func(nums):
    nums.append(4)
    print("함수 안:",nums, id(nums))

data = [1, 2, 3]

func(data)

print("함수 밖:", data, id(data))
# 가변데이터니까 함수 안에서 append()로 4를 추가하면 함수 밖의 data도 변경된다.

# 함수 안: [1, 2, 3, 4] 2514266949248
# 함수 밖: [1, 2, 3, 4] 2514266949248
# id()값이 동일하다.
```

```python
#두 시퀀스의 교집합을 구하는 함수

def my_intersect(a, b) -> list: # 반환타입 명시
    """
    a: 리스트
    b: 리스트
    반환타입: 리스트
    내용: 두 리스트의 중복되는 값만을 새로운 리스트로 반환
    """
    result = []
    for e in a:
        if e in b:
            result.append(e)
    return result

print(my_intersect([1, 2, 3], [2, 3, 5]))
print(my_intersect((1, 2, 3), (2, 3, 5)))
print(my_intersect({1, 2, 3}, {2, 3, 5}))
print(my_intersect({"id":"abc", "sn":"1111"}, {"id":"xyz", "sn":"1234"})) 
```

```python
# 두 시퀀스의 공통 요소 개수

def my_count_common(a, b):
    count = 0
    for e in a:
        if e in b:
            count += 1
    return count


print(my_count_common([1, 2, 3], [2, 3, 5]))
print(my_count_common((1, 2, 3), (2, 3, 5)))
print(my_count_common({1, 2, 3}, {2, 3, 5}))
print(my_count_common({"id":"abc", "sn":"1111"}, {"id":"xyz", "pw":"1234"}))
```

```python
# a에는 있고 b에는 없는 요소 (차집합)

def my_difference(a, b):
    result = []
    for e in a:
        if e not in b:
            result.append(e)
    return result


print(my_difference([1, 2, 3], [2, 3, 5]))
print(my_difference((1, 2, 3), (2, 3, 5)))
print(my_difference({1, 2, 3}, {2, 3, 5}))
print(my_difference({"id":"abc", "sn":"1111"}, {"id":"xyz", "pw":"1234"}))
```

```python
# 두 시퀀스의 합집합

def my_union(a, b):
    result = []

    for e in a:
        if e not in result:
            result.append(e)

    for e in b:
        if e not in result:
            result.append(e)

    return result


print(my_union([1, 2, 3], [2, 3, 5]))
print(my_union((1, 2, 3), (2, 3, 5)))
print(my_union({1, 2, 3}, {2, 3, 5}))
```

```python
# value가 시퀀스 안에 있는지 확인

def my_contains(data, value):

    for e in data:
        if e == value:
            return True # return은 함수 종료와 동시에 호출한 곳으로 값을 반환한다.

    return False


print(my_contains([1, 2, 3], 2))
print(my_contains(("a", "b", "c"), "d"))
print(my_contains({"id":"abc", "pw":"1234"}, "id"))
```

```python
# 특정 값이 몇 번 나오는지 계산

def my_count(data, target):

    count = 0

    for e in data:
        if e == target:
            count += 1

    return count


print(my_count([1, 2, 1, 3, 1], 1))
print(my_count(("a", "b", "a"), "a"))
print(my_count("banana", "a"))
```

```python
# 중복 제거

def my_unique(data):

    result = []

    for e in data:
        if e not in result: # set()을 사용하지 않고 중복 제거
            result.append(e)

    return result


print(my_unique([1, 2, 2, 3, 1, 4]))
print(my_unique(("A", "A", "B", "C")))
print(my_unique("banana"))
```
