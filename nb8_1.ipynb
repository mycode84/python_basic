```python
greet = "Hello"
while greet: # 문자열의 경우 공백이면 False, 아니면 True
    print(greet)
    greet = greet[1:]
```

```python
a, b = 0, 10 
# 다중 할당 
# 내부적으로 오른쪽의 튜플 (0, 10)을 왼쪽의 변수들에 언패킹해서 각각 저장

while a < b:
    print(a)
    a = a + 1 #a += 1
```

```python
# for는 while처럼 조건문이 들어오는 것이 아닌 자료구조(컬렉션)이 들어온다.

for a in [5, 10, 15]:
    print(a)

for (a, b) in [(1, 2), (10, 20)]:
    print(a)
    print(b)

for a, *b, c in  [(1, 2, 3, 4), (5, 6, 7, 8)]:
    print(a, b, c) # *은 남아있는 변수를 리스트로 받음
```

```python
# break가 없는 경우

for i in range(3):
    print(i)
else:
    print("정상 종료")
```

```python
# break가 있는 경우

for i in range(3):
    if i == 1:
        break
    print(i)
else:
    print("정상 종료") # break가 실행되었으므로 else는 실행되지 않는다.
```

```python
i = 1

while i <= 5:
    if i == 3:
        break
    print(i)
    i += 1

print("반복 종료")
```

```python
try:
    # 예외가 발생할 수 있는 코드

except 예외종류:
    # 예외 처리

else:
    # 예외가 없을 때 실행 (선택)

finally:
    # 항상 실행 (선택)
```

```python
try:
    num = int(input("숫자 입력 : ")) ** 2
    print(num)
except:
    print("숫자가 아닙니다.") # 문자 넣었을 때 오류 뜨는 것을 방지

print("다음 실행문...")
```

```python
try:
    num = int(input("숫자 입력 : "))
    result = 10 / num # 파이썬에서는 무한대를 가질 수 없음
    print(result)
    print("자원해지")

except ValueError:
    print("숫자를 입력하세요.")

except ZeroDivisionError:
    print("0으로 나눌 수 없습니다.")

finally: # 항상 실행. 예외가 발생해도 실행 / else는 예외가 없을 때 실행
    print("프로그램 종료")
```

```python
try:
    print(10 / 2)
except:
    print("오류 발생")
finally:
    print("프로그램 종료")
```

```python
try
 │
 ├── 예외 발생 ─────► except
 │
 └── 예외 없음 ─────► else
          │
          ▼
       finally (항상 실행)
```
