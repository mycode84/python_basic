```python
a = 5
b = 2

print(a + b)   # 7
print(a // b)  # 2 몫
print(a % b)   # 1 나머지
print(5 ** 2)
```

```python
print(5 > 3)      # True
print(5 == 3)     # False
```

```python
print(True and False)   # False
print(True or False)    # True
print(not True)         # False

print(10>5 and 10==7)
```

```python
print(3 in [1, 2, 3])       # True
a = [1, 2, 3]
print(4 in a)  # False
print(4 not in a)   # True
```

```python
a = [1, 2]
b = a

a is b      # True
a == b      # True
```

```python
# 삼항연산자
score = 50
result = "합격" if score >= 60 else "불합격"
print(result)
```

```python
a = 10

# a = a + 5와 같은 표현(편한거 사용하면 된다) 
a += 5 # 헷갈리니까 a = a + 5 형식으로 사용하기!
print(a)  # 15
```

```python
n = 10
result = (x := n>5)   # 대입하면서 값 / x = n>5 의 형태는 문이기 때문에 값을 반환하지 않아서 result에 대입할 수 없음
print(result)
```
