```python
# dict도 순서를 가지고 있다. (호출이 아닌 반복을 위해서 / set은 순서를 가지지 않는다.)

# 1. 표현식 활용
student = {
    "name": "홍길동",
    "age": 20,
    "score": 90
}

print(student)
print(type(student))
```

```python
# 2. dict() 함수 활용

person = dict(name="김철수", age=30, city="서울")

print(person)


# 3. [()]를 dict으로 변환
data = [
    ("apple", 1000),
    ("banana", 2000),
    ("orange", 3000)
]

fruit = dict(data)

print(fruit)
```

```python
d = {"name": "홍길동"}
print(d)

d = {100: "학생"}
print(d)

d = {
    (1, 2): "좌표"
}
print(d)

# d = {
     # [1, 2]: "리스트"
# } # error 리스트는 가변타입이기 때문에 키값으로 가질 수 없다. (튜플은 불변타입이기 때문에 키값으로 가질 수 있다.)
print(d)
```

```python
student = {"name": "홍길동"}
print(student)

# 요소조회
print(student["name"])

# 추가
student["age"] = 20
print(student)

# 수정
student["age"] = 30
print(student)

# 삭제 del문 : 반환값이 없다
del student["age"]
print(student)

# in 연산
print("name" in student)
print("age" in student)
```

```python
student = {"name": "홍길동"}

print(student.get("name"))
print(student.get("age")) # 없으면 None 반환

student = {"name": "홍길동", "age": 20}
print(student.keys())
print(student.values())
print(student.items())

student.update({"age": 30, "score": 90})
print(student)

age = student.pop("age")
print(age)
print(student)

student.clear()
print(student)
```

```python
# for 반복문 안에는 in 연산자가 포함되어 있다.

fruits = ["apple", "banana", "orange"]

for fruit in fruits:
    print(fruit)
```

```python
colors = ("red", "green", "blue")

for color in colors:
    print(color)
```

```python
nums = {10, 20, 30} # set은 순서를 가지지 않는다. 하지만 임의로 순서를 정해 출력할 수 있다.

for num in nums:
    print(num)
```

```python
student = {
    "name": "홍길동",
    "age": 20
}
# 키
for key in student:
    print(key)
# 값
for value in student.values():
    print(value)

# 키 + 값 # item() 키 값을 하나의 튜플로 묶어서 반환 -> 언패킹 키 따로, 값 따로
for key, value in student.items():
    print(key, value)

for key, *value in student.items():
    print(key, value)  # *value를 사용하면 value가 리스트로 묶여서 반환된다. (튜플이 아닌 리스트로 반환됨)
```
