```python
n = 10
print(n.bit_length()) # 10의 이진수 값(1010)의 길이

# bin()함수는 인자로 입력된 10진수의 2진수 문자열을 반환한다.
# 이진수 문자열 앞에 binary를 뜻하는 "0b"접두사를 앞에 붙여서 반환한다. (1010과 10, 10을 구분하기 위함)
print(bin(n)) #0b1010 "0b" + "1010"

# bit_length()메서드는 n이라는 숫자 객체를 만들어야 존재하는 것이고,
# bin() 함수는 n객체와 상관없는 독립된 함수 객체이다.

print(bin(9)) # n과 상관없이 독립적으로 사용가능
```

```python
nums = [10, 20, 30, 40, 50]

print(nums[1:4]) # [20, 30, 40]
print(nums[:3]) # [10, 20, 30]
print(nums[2:]) # [30, 40, 50]
print(nums) #  [10, 20, 30, 40, 50] 슬라이싱 연산 후에도 리스트 자신은 변경되지 않는다. 

# 2칸씩 건너뛰기
nums = [10, 20, 30, 40, 50, 60, 70, 80]
print(nums[::2])

nums = [10, 20, 30, 40, 50, 60, 70, 80]
print(nums[1::2])

# 역순 출력
nums = [10, 20, 30, 40, 50]
print(nums[::-1]) # [50, 40, 30, 20, 10]

# 뒤에서 3개 추출
nums = [10, 20, 30, 40, 50, 60]
print(nums[-3:])

# 뒤에서 4개 중 2칸씩
nums = [10, 20, 30, 40, 50, 60, 70, 80]
print(nums[-4::2])

# 뒤에서 앞으로 2칸씩
nums = [10, 20, 30, 40, 50, 60, 70, 80]
print(nums[::-2])

# [5]위치부터 뒤로(-1) [1]위치 전까지
nums = [10, 20, 30, 40, 50, 60, 70]
print(nums[5:1:-1]) # 5위치부터 1전까지 -1칸식  [60, 50, 40, 30]

# 마지막 5개를 역순으로
nums = [10, 20, 30, 40, 50, 60, 70, 80]
print(nums[-1:-6:-1])

# 슬라이싱 할당(슬라이싱 구간에 대입문)
nums = [10, 20, 30, 40, 50]
nums[1:4] = [7, 8] # 1~3구간 [20, 30, 40]에 [7, 8]을 대입 
print(nums) # [10, 7, 8, 50]

# 슬라이싱 결과를 다른 변수에 대입
a = [10, 20, 30, 40, 50, 60]
b = a[:]
c = a[::-1]
d = a[1::2]
e = a[:2]

print(b)
print(c)
print(d)
print(e)
```

```python
nums = [50, 10, 40, 20]
n2 = nums.sort(reverse=True) # 내림차순
print(nums)
print(n2) # 정렬만 하고 반환을 하지 않으니까 값이 None으로 나온다. sort()메서드는 리스트 객체를 정렬만 하고 반환하지 않는다.
```

```python
# insert() : 원하는 위치에 추가
fruits = ["apple", "banana"]
fruits.insert(1, "orange")
print(fruits)


# extend() : 여러 개 추가
nums = [1, 2, 3]
nums.extend([4, 5, 6])
print(nums)


# remove() : 값으로 삭제
fruits = ["apple", "banana", "orange"]
fruits.remove("banana")
print(fruits)


# 수정은 인덱스 할당, 슬라이싱 할당으로 수정
# pop() : 인덱스로 삭제 후 반환
nums = [10, 20, 30]
result = nums.pop() # 마지막 인덱스 삭제 후 반환
print(result)
print(nums)

nums = [10, 20, 30]
result = nums.pop(1) # 인덱스1 삭제 후 반환
print(result)
print(nums)


# clear() : 전체 삭제
nums = [10, 20, 30]
nums.clear()
print(nums)
```

```python
# 튜플은 팩킹과 언팩킹!!!

t = 1, 3, 5 # 팩킹 실행 -> 리스트가 아닌 튜플로 묶어짐
print(t) 
print(type(t)) # <class 'tuple'>

x, y, z = t # 언팩킹 실행
print(x, y, z)

a, b = [3,6] # (3,6) 팩킹 발생
print(a, b)
```

```python
# 셋은 집합의 개념 / 중복이 허용되지 않는다 / 순서가 없다 
a = {1, 2, 3, 2}
print(a)
# print(a[0]) # set은 인덱스가 없다
print(list(a)[0]) # set을 list로 변환 후 인덱스로 접근 가능
```

```python
a = {"1": "admin", "pw": 1234} # 딕셔너리
print(a, type(a))

b = {3,4,6} # 셋
print(b, type(b))

c = []
print(type(c))

d = {} # 빈 중괄호는 딕셔너리와 셋 중 딕셔너리이다
print(type(d))

e = set() # 빈 set()은 셋이다
print(type(e))
```

```python
# 집합 연산은 기호를 사용하거나, 메서드를 사용할 수 있는데 / 기호만으론 가독성이 떨어지니까, 메서드 사용을 권하심
a = {1, 2, 3}
b = {3, 4, 5}

# 교집합
print(a & b) 
print(a.intersection(b))

# 합집합
print(a | b)
print(a.union(b))

# 차집합
print(a - b)
print(a.difference(b))
```

```python
# 패킹, 언패킹시 * 연산자 : 나머지 값들은 하나씩 가져가고, * 연산자를 붙인 변수에는 나머지 값들을 리스트(가변객체)로 묶어서 가져간다.
# a, b = [1, 2, 3, 4] 

*a, b = (1, 2, 3, 4)
print(a)
print(b)

c, *d, e = (1, 2, 3, 4)
print(c)
print(d)
print(e)

f = [1, 2, 3]
print(*f) # 1 2 3 # print(h[0], h[1], h[2])와 동일한 결과
```
