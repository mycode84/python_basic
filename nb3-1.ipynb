```python
# 2. 부동 소수점 수 (float) : 오차를 가진다 (소수점 16번째까지만 정확도를 가짐)
b = 3.14
print(b)
print(type(b))

print(0.1+0.2)
```

```python
# 3. 복소수 (complex)
c = 3 + 4j
print(c)
print(type(c))
```

```python
# 4. 논리값 (bool)
d = True
print(d)
print(type(d))
print(3>5)
```

```python
# 5. 문자열 (str)
e = "Python"
print(e)
print(type(e))

# 인덱싱
print(e[1])

# e[1] = "k" # error 문자열객체 불변이다
```

```python
# 6. 리스트 (list)
f = [1, 2, 3]
print(f)
print(type(f))

f[1] = "abc"
print(f)
```

```python
# 7. 튜플 (tuple) # 리스트와 달리 값을 변경할 수 없는 불변객체
g = (1, 2, 3)
print(g)
print(type(g))
```

```python
# 8. 딕셔너리 (dict) # 키에는 뭐든지 들어올 수 있고, 값은 바꿀 수 있다(가변)
h = {"id" : "admin", "pw" : "1234"}
print(h)
print(type(h))

print(h["id"])
h["pw"] = "0000"
print(h)
```

```python
# 9. 집합 (set) # 집합 안에 중복된 값은 있을 수 없다
i = {1, 2, 3}
print(i)
print(type(i))
```

```python
# 10. 범위 (range)
j = range(5) 
print(j)
print(type(j))
print(list(j))
```
