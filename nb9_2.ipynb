```python
# <람다함수 문법>

# 함수 정의는 함수선언문 이다 -> 함수를 정의하는 문

# 람다는 함수 표현식 이다 -> 람다 표현식의 결과로 함수를 반환한다.

# 람다 자체는 함수(객체)를 반환하고 람다를 호출하면 람다의 반환값이 반환된다.

# lambda 매개변수: 반환값
```

```python
# 일반 함수 정의 # add1이라는 함수를 생성
def add1(a, b):
    return a + b

# 람다 함수 # 람다를 익명함수로 부르기도 함 # 변수저장 -> 1급 함수 기능 포함
add2 = lambda a, b : a + b 
# add2에는 "lambda a, b : a + b" 람다가 반환한 함수(객체)가 대입

r1 = add1(5, 1)
print(r1)

r2 = add2(1, 3)
print(r2)

# 변수를 설정하지 않는 경우 (lambda a, b : a + b)()로도 표현 가능
r3 = (lambda a, b : a + b)(1, 3)
print(r3)
```

```python
# def ds_func(data, func):
    # for d in data:
        # func()



def test(func):
    a = 2
    return func(2) + 1000


def square(x):
    return x * x

def square2(x):
    return x - x

def square3(x):
    return x + x


r1 = test(square)
print(r1)

r2 = test(square2)
print(r2)

r3 = test(square3)
print(r3)


r4 = test(lambda x: x * x) # 매번 특정 함수를 정의하지 않고 람다를 사용하여 간단히 표현 가능
print(r4)
```

```python
def score(item): # 아이템을 입력받아 1번째 요소를 반환 ex)  ('홍길동', 90)가 인수로 입력되면 90 반환
    return item[1]

s = [
    ('홍길동', 90),
    ('이순신', 85),
    ('강감찬', 100)
]

r = sorted(s, key=score, reverse=True) # key 매개변수에 score 함수를 대입, 람다를 사용하여 간단히 표현 가능

print(r)

r2 = sorted(s, key=lambda e: e[1], reverse=False) 
print(r2)
# lambda의 단점: 식이기 때문에 문을 넣을 수 없다. 따라서 단일 컴프리헨션만 가능하다.
```

```python
# 1) map() - 모든 요소를 함수 처리해서 반환

nums = [1, 2, 3, 4]

# 리스트의 각 요소(x)에 2를 곱한 결과를 반환
result = list(map(lambda x: x * 2, nums)) # map을 list로 변환을 해줘야 print로 출력 가능


print(result) #[2, 4, 6, 8]
```

```python
# 2) filter() - 조건에 맞는 요소만 선택 # filter은 True, False를 반환하는 함수가 필요하다.

nums = [1, 2, 3, 4, 5]

# 짝수인 요소만 선택
result = list(filter(lambda x: x % 2 == 0, nums)) # filter를 list로 변환을 해줘야 print로 출력 가능

print(result)
```

```python
# 3) sorted() - 정렬 기준 지정 sorted(data, key=len)

names = ["banana", "apple", "kiwi"]

# 문자열의 길이를 기준으로 오름차순 정렬
result = sorted(names, key=lambda x: len(x)) # 6, 5, 4를 오름차순으로 정렬

print(result)
```

```python
# stored()에 람다를 사용하면 데이터 과학에서 자주 사용되는 [dict, dict, ...] 자료구조를 원하는 키값으로 정렬 가능

students = [
    {"이름": "홍길동", "나이": 20, "점수": 90},
    {"이름": "이순신", "나이": 18, "점수": 85},
    {"이름": "김유신", "나이": 22, "점수": 95}
]

# "나이" 값을 기준으로 오름차순 정렬 
result = sorted(students, key=lambda x: x["나이"]) # 20, 18, 22를 오름차순으로 정렬
print(result)

# x : 리스트의 각 딕셔너리
# x["나이"] : 각 딕셔너리의 "나이" 값을 꺼냄
# sorted()는 이 값을 기준으로 정렬한다.
```

```python
# "점수"를 기준으로 내림차순 정렬
result = sorted(students, key=lambda x: x["점수"], reverse=True)
print(result)

# "이름"을 기준으로 오름차순 정렬
result = sorted(students, key=lambda x: x["이름"])
print(result)
```

```python
# 여러 키를 기준으로 정렬도 가능하다

students = [
    {"이름": "홍길동", "반": 2, "점수": 90},
    {"이름": "이순신", "반": 1, "점수": 90},
    {"이름": "김유신", "반": 1, "점수": 95}
]

# 1순위: 반, 2순위: 점수
result = sorted(students, key=lambda x: (x["반"], x["점수"])) # 하나의 값을 받아서 튜플로 반환
print(result)
```

```python
# 4) min() - 기준에 따라 최소값 찾기

names = ["banana", "apple", "kiwi"]

# 문자열 길이가 가장 짧은 요소 반환
result = min(names, key=lambda x: len(x))

print(result) # kiwi
```

```python
# 5) max() - 기준에 따라 최대값 찾기

names = ["banana", "apple", "kiwi"]

# 문자열 길이가 가장 긴 요소 반환
result = max(names, key=lambda x: len(x))

print(result) # banana
```

```python
# 6) any() - <하나라도> 조건을 만족하는지 확인

nums = [1, 3, 5, 8]

# 각 요소가 짝수인지(True/False)로 변환
# 하나라도 True이면 True 반환


r1 = map(lambda x: x % 2 == 0, nums) 
print("r1 =", r1) 
r2 = list(r1)
print("r2 =", r2) # [False, False, False, True]
result = any(r1)


result = any(map(lambda x: x % 2 == 0, nums)) 
# any()는 True, False를 반환하는 함수가 필요하다. -> map이 가장 적합

print(result) # True
```

```python
# 7) all() - <모두> 조건을 만족하는지 확인

nums = [2, 4, 6, 8]

# 각 요소가 짝수인지(True/False)로 변환
# 모두 True이면 True 반환
result = all(map(lambda x: x % 2 == 0, nums))

print(result) #True
```

```python
# 2. 함수의 인수로 전달하기

def square(x):
    return x * x

def test(func): # 함수 속 함수: 고차함수
    return func(5)

print(test(square))
```

```python
# 3. 함수를 반환하기

def outer():
    def inner():
        print("inner")

    return inner

f = outer()

f()

(outer())() # outer()가 반환한 inner()를 호출
```

```python
# 4. 자료 구조에 저장하기 # 딕셔너리 안에 함수 저장

def plus(x):
    return x + 10

def minus(x):
    return x - 10

funcs = [plus, minus]

print(funcs[0](100))
print(funcs[1](100))


def add(x, y):
    return x + y

def sub(x, y):
    return x - y

calc = {
    "+": add,
    "-": sub
}

print(calc["+"](10, 5))
print(calc["-"](10, 5))
```

```python
# zip()
# zip()은 여러 개의 시퀀스 객체를 같은 위치(인덱스)의 값끼리 묶어주는 함수이다.

# zip -> list

names = ["홍길동", "이순신", "김유신"]
scores = [90, 80, 100]
result = zip(names, scores)
print(list(result)) # [('홍길동', 90), ('이순신', 80), ('김유신', 100)]

names = ["홍길동", "이순신", "김유신"]
scores = [90, 80, 100]
grades = ["A", "B", "A"]

print(list(zip(names, scores, grades)))
```

```python
# 반복문에서 사용

names = ["홍길동", "이순신", "김유신"]
scores = [90, 80, 100]

for name, score in zip(names, scores):
    print(name, score)

# 홍길동 90
# 이순신 80
# 김유신 100
```

```python
# zip -> dict

names = ["홍길동", "이순신", "김유신"]
scores = [90, 80, 100]

result = dict(zip(names, scores))

print(result) # {'홍길동': 90, '이순신': 80, '김유신': 100}
names = ["홍길동", "이순신", "김유신"]
scores = [90, 80, 100]

result = {name: score for name, score in zip(names, scores)}

print(result)
```
