```python
nums = range(1, 11)
list_nums = list(nums)
print(list_nums)

even = []
for i in list_nums:
    if i%2 == 0:
        even.append(i)
print(even)

even2 = [i for i in list_nums if i%2 == 0] # 식이기 때문에 :는 생략 (=컴프리헨션)
print(even2)

even3 = [i * 10 for i in list_nums if i%2 == 0] 
print(even3)

names = ["미치에다슌스케", "혼다쿄야", "타카"] # n[:1], n[1:]

print([n+"님" for n in names if n.startswith("타")])

names2 = ["야기유세이", "혼다쿄야", "스다마사키"]
print({n[:2]:n[2:] for n in names2})
```
