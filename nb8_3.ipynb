```python
import csv

with open("score.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)

    data = list(reader)
    for row in data:
        print(row)

# 결측치 & 이상치 처리 # 나이
sum_age = 0
count = 0

for d in data:
    age = d["나이"]

    if age != "":
        age = int(age)

        if age <= 100:
            sum_age += age
            count += 1

avg_age = int(sum_age / count)
print(avg_age)
            
for d in data:
    age = d["나이"]
    
    if age == "":
        d["나이"] = avg_age
    else:
        age = int(age)

        if age >= 100:
           d["나이"] = avg_age

data

# 결측치 & 이상치 처리 # 국어

sum_language = 0
count2 = 0

for d in data:
    language = d["국어"]

    if language != "결측" and language != "":
        language = int(language)

        if language <= 100 and language > 0:
            sum_language += language
            count2 += 1

avg_language = int(sum_language / count2)
print(avg_language)
            

for d in data:
    language = d["국어"]

    if language == "결측" or language == "":
        d["국어"] = avg_language
    else:
        language = int(language)

        if language > 100 or language < 0:
           d["국어"] = avg_language 

data

# 결측치 & 이상치 처리 # 영어

sum_english = 0
count3 = 0

for d in data:
    english = d["영어"]

    if english != "NaN" and english != "":
        english = int(english)

        if english <= 100 and english > 0:
            sum_english += english
            count3 += 1

avg_english = int(sum_english / count3)
print(avg_english)
            

for d in data:
    english = d["영어"]

    if english == "NaN" or english == "":
        d["영어"] = avg_english
    else:
        english = int(english)

        if english > 100 or english < 0:
           d["영어"] = avg_english

data



# 파생속성 - 총점 & 평균

for d in data:
    language = int(d["국어"])
    english = int(d["영어"])

    total = language + english

    d["총점"] = total

avg = (language + english)/2

for d in data:
    d["평균"] = avg

print(data)



# 특성추출

for d in data:
    avg = d["총점"] / 2

    if avg >= 60:
        d["결과"] = "합격"
    else:
        d["결과"] = "불합격"

print(data)
```
