```python
# try-finally는 파일을 수동으로 닫는 방법
try:
    f = open("students.csv", "r", encoding="utf-8")
    print(f.read())
    print(10 / 0)
finally: # 예외가 발생하더라도 실행되게끔
    f.close() 
    print("자원해지 성공")
```

```python
# with는 파일을 자동으로 닫아주는 문법
with open("students.txt", "r", encoding="utf-8") as f:
    print(f.read())
    print(10 / 0)
```

```python
with open("student.txt", "r", encoding="utf-8") as f:
    for line in f:
        print(line.strip()) # line에 양쪽 공백제거(줄바꿈포함)하여 print함수의 기본 줄바꿈만으로 출력
```

```python
# 한 라인을 구분자","를 이용하여 두개의 문자열 값으로 저장

names = [] # 리스트로 저장
scores = []
with open("student.txt", "r", encoding="utf-8") as f:
    for line in f:
        name, score = line.strip().split(",") # 공백제거, "," 로 구분
        print(name, int(score))
        names.append(name)
        scores.append(score)
print(names)
print(scores)
```

```python
students = {} # 딕셔너리로 저장

with open("student.txt", "r", encoding="utf-8") as f:
    for line in f:
        name, score = line.strip().split(",")
        students[name] = int(score)

print(students)
```
