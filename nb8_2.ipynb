```python
# try-finally는 파일을 수동으로 닫는 방법
try:
    f = open("students.csv", "r", encoding="utf-8")
    print(f.read())
    print(10 / 0)
finally: # 예외가 발생하더라도 실행되게끔
    f.close() 
    print("자원해지 성공")
```

```python
# with는 파일을 자동으로 닫아주는 문법
with open("students.txt", "r", encoding="utf-8") as f:
    print(f.read())
    print(10 / 0)
```

```python
with open("student.txt", "r", encoding="utf-8") as f:
    for line in f:
        print(line.strip()) # line에 양쪽 공백제거(줄바꿈포함)하여 print함수의 기본 줄바꿈만으로 출력
```

```python
# 한 라인을 구분자","를 이용하여 두개의 문자열 값으로 저장

names = [] # 리스트 자료구조화
scores = []
with open("student.txt", "r", encoding="utf-8") as f:
    for line in f:
        name, score = line.strip().split(",") # 공백제거, "," 로 구분
        print(name, int(score))
        names.append(name)
        scores.append(score)
print(names)
print(scores) 
```

```python
students = {} # 딕셔너리 자료구조화

with open("student.txt", "r", encoding="utf-8") as f:
    for line in f:
        name, score = line.strip().split(",")
        students[name] = int(score)

print(students) 
```

```python
import csv

with open("students.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f) # 헤드라인을 키, 한 줄의 내용을 값으로 딕셔너리 생성

    for row in reader:
        print(row)
```

```python
# csv모듈의 reader반환 타입은 읽기 속도는 빠르지만(iterator), 자료구조 내부를 핸들링하는데는 여러가지 제한 O
# 필요하다면 list(reader)로 리스트로 만들어 인덱싱, 슬라이싱, len() 등을 사용.
# 큰 CSV 파일은 reader 반복으로 처리하고, 작은 CSV 파일은 list(reader)로 메모리에 올려 사용.

import csv

with open("students.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)

    data = list(reader)

print(data)
data[2]["영어"] = 100 # 리스트의 형태로 불러왔기 때문에 수정 가능 / csv 원본이 수정되는 것은 아님
print(data[2])

# 변경한 csv 파일을 저장하고 싶은 경우 "w" 를 사용
```

```python
# csv.DictReader()로 읽어서 data를 수정했다면, csv.DictWriter() 또는 일반 파일 쓰기를 이용해 다시 저장

import csv

# students_result.csv 파일을 쓰기(w) 모드로 열기
# newline="" : 빈 줄이 추가되는 현상 방지
# encoding="utf-8-sig" : 한글이 엑셀에서 깨지지 않도록 저장
with open("students_result.csv", "w", newline="", encoding="utf-8-sig") as f:
    # data[0].keys() → 첫 번째 딕셔너리의 키(컬럼명) 사용    
    writer = csv.DictWriter(f, fieldnames=data[0].keys())
    # 헤더(컬럼명) 저장
    writer.writeheader()   
    # data 리스트에 있는 모든 딕셔너리를 한 번에 저장
    writer.writerows(data)  

print("CSV 저장 완료")
```
